```python
import pandas as pd
import json
```

```python
with open("../data/id_to_remove.json", "r") as f:
    id_to_remove = set(json.load(f))
```

```python
df = pd.read_csv("../data/hotels.tsv", sep="\t")
```

```python
from collections import Counter
c = Counter()
for id, review_type in zip(df['id'], df['review_type']):
    if id in id_to_remove: continue
    c[review_type] += 1
```
